```python
with open('wiki_infobox_person.txt', 'r') as f:
    lines = f.readlines()
```

```python
lines
```

```python
import re
import pandas as pd
```

```python
def clean_line(l):
    return re.sub(' \(Transclusion count: [0-9,]+\)( Holding cell|)', '', l).strip().replace('Template:', '')
def get_count(l):
    return int(re.search('[0-9,]+', l)[0].replace(',', ''))
```

```python
lst_infobox = []
belongs = ['', '', '', '', '']
for l in lines:
    cl = clean_line(l)
#     if len(re.findall(' {4}', l)) == 0:
#         belongs[0] == cl
#     elif len(re.findall(' {4}', l)) == 1:
#         belongs[1] == cl
#     elif len(re.findall(' {4}', l)) == 2:
#         belongs[2] == cl
#     elif len(re.findall(' {4}', l)) == 3:
#         belongs[3] == cl
    belongs[len(re.findall(' {4}', l))] = cl
    if re.search('Template:Infobox ', l) is not None:
        info = [cl, get_count(l)] + belongs
        info = info[:info[1:].index(info[0]) + 1]
        lst_infobox.append(info)
```

```python
df_infobox = pd.DataFrame(lst_infobox, columns=['infobox_type', 'count', 'belong_to_1', 'belong_to_2', 'belong_to_3', 'belong_to_4'])
```

```python
df_infobox.to_csv('infobox.csv', sep=',', index=False)
```
